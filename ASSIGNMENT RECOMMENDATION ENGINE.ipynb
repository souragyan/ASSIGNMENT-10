```python
import chardet
import pandas as pd

with open('finalb.csv', 'rb') as f:
    result = chardet.detect(f.read())  # or readline if the file is large


df=pd.read_csv('finalb.csv', encoding=result['encoding'])
df
```

```python
booknew=df.rename({'User.ID':'id','Book.Title':'title','Book.Rating':'rating'},axis=1)
```

```python
booknew[0:50].sort_values(by=['id'],ascending=False)
```

```python
#number of unique users in the dataset
len(booknew.id.unique())
```

```python
len(booknew.title.unique())
```

```python
user_book= booknew.pivot_table(index='id',columns='title',values='rating')
user_book
```

```python
user_book.fillna(0, inplace=True)
```

```python
user_book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 - pairwise_distances(user_book.values,metric='cosine')
```

```python
user_sim
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = booknew.id.unique()
user_sim_df.columns = booknew.id.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```
